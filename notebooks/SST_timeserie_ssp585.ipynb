## Computing the timeserie

We're going to reduce the data to spatially averaged value (time varying). Since the area of the ocean cells is not provided,
we need to compute them. Fortunately it is doable analytically on a regularly spaced grid:

```python
def add_areacello_regular_grid(ds, Rearth=6378e3):
    """ compute the cells area on a regular grid """

    rfac = 2 * np.pi * Rearth / 360

    up = {"bnds": 1}
    down = {"bnds": 0}

    dx1d = rfac * (ds["lon_bnds"].isel(up) - ds["lon_bnds"].isel(down))
    dy1d = rfac * (ds["lat_bnds"].isel(up) - ds["lat_bnds"].isel(down))

    dx2d, dy2d = np.meshgrid(dx1d, dy1d)
    _, lat2d = np.meshgrid(ds["lon"].values, ds["lat"].values)

    dx = dx2d * np.cos(2 * np.pi * lat2d / 360)
    dy = dy2d
    area = dx * dy
    
    ds['areacello'] = xr.DataArray(data=area, dims=('lat', 'lon'))
    return ds
```

```python
CM4_ssp585 = add_areacello_regular_grid(CM4_ssp585)
```

```python
CM4_ssp585
```

Now let's compute the spatial average:

```python
def avg2dfield(ds, field):
    """ compute a 2d masked average """
    area_masked = ds['areacello'].where(~np.isnan(ds[field].isel(time=0)))
    avg = (ds[field] * area_masked).sum(dim=['lon', 'lat']) / (area_masked).sum(dim=['lon', 'lat'])
    return avg
```

Wow this is done incredibly fast! But there is a catch: all of this is done lazily! no computations have been made yet, only the object that will contain the data is created:

```python
mean_tos_CM4 = avg2dfield(CM4_ssp585, 'tos')
```

If we want the data, we need to query for values (e.g. make a plot). This computation is intensive so to be able to perform it, we're gonna spin a dask compute cluster:

```python
from dask.distributed import Client
from dask_gateway import Gateway


gateway = Gateway()  # connect to Gateway

cluster = gateway.new_cluster()  # create cluster
cluster.scale(10)  # scale cluster

client = Client(cluster)  # connect Client to Cluster
```
